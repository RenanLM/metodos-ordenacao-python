```python
# Bubble Sort de Renan Lucas de Moura

import numpy as np
import matplotlib.pyplot as plt
from random import randint
import time
```

Função Bubble Sort

```python
def bubble_sort(lista):
    for i in range(len(lista)):
        flag = False  # Flag para monitorar se houve troca entre um maior e menor
        for j in range(len(lista) - 1 - i):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                flag = True
        if not flag:
            break
    return lista
```

Tamanhos das listas e tempos, melhor, pior e caso médio(aleatório)

```python
tamanhos = [1000, 2000, 4000, 6000, 8000, 10000]

tempo_melhor_caso = []
tempo_pior_caso = []
tempo_caso_aleatorio = []
```

Geração das listas e cálculos de tempos

```python
for tamanho in tamanhos:
    # melhor caso: lista já ordenada

    lista_melhor = list(range(tamanho))
    inicio = time.perf_counter()
    bubble_sort(lista_melhor.copy())
    fim = time.perf_counter()
    tempo_melhor_caso.append(fim - inicio)

    # pior caso: lista em ordem decrescente

    lista_pior = list(range(tamanho, 0, -1))
    inicio = time.perf_counter()
    bubble_sort(lista_pior.copy())
    fim = time.perf_counter()
    tempo_pior_caso.append(fim - inicio)

    # caso médio: ordem aleatório

    lista_aleatoria = np.random.randint(0, 100000, tamanho)
    inicio = time.perf_counter()
    bubble_sort(lista_aleatoria.copy())
    fim = time.perf_counter()
    tempo_caso_aleatorio.append(fim - inicio)
```

Gráficos comparando os tempos

```python
plt.plot(tamanhos, tempo_melhor_caso, marker='o', linestyle='-', color='green', label='Melhor caso')
plt.plot(tamanhos, tempo_pior_caso, marker='o', linestyle='-', color='red', label='Pior caso')
plt.plot(tamanhos, tempo_caso_aleatorio, marker='o', linestyle='-', color='blue', label='Caso aleatório')
plt.title('Tempo de Execução do Bubble Sort em Diferentes Casos')
plt.xlabel('Tamanho da Lista')
plt.ylabel('Tempo de Execução (segundos)')
plt.legend()
plt.grid(True)
plt.show()
```
