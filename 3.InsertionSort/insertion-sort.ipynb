```python
#Insertion Sort de Renan Lucas de Moura

import numpy as np
import matplotlib.pyplot as plt
from random import randint
import time
```

Função Insertion Sort

```python
def insertion_sort(lista):
  n = len(lista)
  if n <= 1:
    return

  for i in range(1, n):
    key = lista[i]
    j = i-1
    while j >= 0 and key < lista[j]:
      lista[j+1] = lista[j]
      j -= 1
    lista[j+1] = key
```

Funções Bubble e Selection Sort (para efeitos de comparação)

```python
def bubble_sort(lista):
    for i in range(len(lista)):
        flag = False  # Flag para monitorar se houve troca entre um maior e menor
        for j in range(len(lista) - 1 - i):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                flag = True
        if not flag:
            break
    return lista

def selection_sort(lista):
    for i in range(len(lista)):
        pivo = i
        for j in range(i + 1, len(lista)):
            if lista[j] < lista[pivo]:
                pivo = j
        lista[i], lista[pivo] = lista[pivo], lista[i]
    return lista
```

Tamanhos das listas e tempos, melhor, pior e caso médio(aleatório)

```python
tamanhos = [1000, 2000, 4000, 6000, 8000, 10000]

tempo_melhor_caso = []
tempo_pior_caso = []
tempo_caso_aleatorio = []
tempo_caso_aleatorio_ss = []
tempo_caso_aleatorio_bs = []
```

Geração das listas e cálculos de tempo

```python
for tamanho in tamanhos:
    # melhor caso: lista já ordenada
    lista_melhor = list(range(tamanho))
    inicio = time.perf_counter()
    insertion_sort(lista_melhor.copy())
    fim = time.perf_counter()
    tempo_melhor_caso.append(fim - inicio)

    # pior caso: lista em ordem decrescente
    lista_pior = list(range(tamanho, 0, -1))
    inicio = time.perf_counter()
    insertion_sort(lista_pior.copy())
    fim = time.perf_counter()
    tempo_pior_caso.append(fim - inicio)

    # caso médio: ordem aleatória
    lista_aleatoria = np.random.randint(0, 100000, tamanho)
    inicio = time.perf_counter()
    insertion_sort(lista_aleatoria.copy())
    fim = time.perf_counter()
    tempo_caso_aleatorio.append(fim - inicio)

    # caso médio: ordem aleatória SelectionSort
    lista_aleatoria = np.random.randint(0, 100000, tamanho)
    inicio = time.perf_counter()
    selection_sort(lista_aleatoria.copy())
    fim = time.perf_counter()
    tempo_caso_aleatorio_ss.append(fim - inicio)

    # caso médio: ordem aleatório BubbleSort

    lista_aleatoria = np.random.randint(0, 100000, tamanho)
    inicio = time.perf_counter()
    bubble_sort(lista_aleatoria.copy())
    fim = time.perf_counter()
    tempo_caso_aleatorio_bs.append(fim - inicio)
```

Gráfico Insertion Sort em diferentes casos

```python
plt.plot(tamanhos, tempo_melhor_caso, marker='o', linestyle='-', color='green', label='Melhor caso - Insertion Sort')
plt.plot(tamanhos, tempo_pior_caso, marker='o', linestyle='-', color='red', label='Pior caso - Insertion Sort')
plt.plot(tamanhos, tempo_caso_aleatorio, marker='o', linestyle='-', color='blue', label='Caso aleatório - Insertion Sort')
plt.title('Tempo de Execução do Insertion Sort em Diferentes Casos')
plt.xlabel('Tamanho da Lista')
plt.ylabel('Tempo de Execução (segundos)')
plt.legend()
plt.grid(True)
plt.show()
```

Gráfico Insertion x Bubble x Selection

```python
plt.plot(tamanhos, tempo_caso_aleatorio, marker='o', linestyle='-', color='red', label='Caso aleatório - Insertion Sort')
plt.plot(tamanhos, tempo_caso_aleatorio_ss, marker='o', linestyle='-', color='blue', label='Caso aleatório - Selection Sort')
plt.plot(tamanhos, tempo_caso_aleatorio_bs, marker='o', linestyle='-', color='orange', label='Caso aleatório - Bubble Sort')
plt.title('Comparação do Tempo de Execução (Caso Aleatório)')
plt.xlabel('Tamanho da Lista')
plt.ylabel('Tempo de Execução (segundos)')
plt.legend()
plt.grid(True)
plt.show()
```
